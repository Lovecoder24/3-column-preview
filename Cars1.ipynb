####QUESTION 13

Which manufacturer has the highest and lowest average sales volume?

```python
#calculate the mean of the 'Sales_in_thousands' column (replace 'Sales_in_thousands' if needed)
avg_sales_by_manufacturer = df.groupby('Manufacturer')['Sales_in_thousands'].mean()


# Get the manufacturer with the highest average sales
highest_avg_sales = avg_sales_by_manufacturer.idxmax()
highest_avg_sales_value = avg_sales_by_manufacturer.max()

# Get the manufacturer with the lowest average sales
lowest_avg_sales = avg_sales_by_manufacturer.idxmin()
lowest_avg_sales_value = avg_sales_by_manufacturer.min()

# Display the results
print(f"Manufacturer with the highest average sales volume: {highest_avg_sales} ({highest_avg_sales_value})")
print(f"Manufacturer with the lowest average sales volume: {lowest_avg_sales} ({lowest_avg_sales_value})")
```

####QUESTION 14

What is the distribution of car prices in the dataset?

```python


import matplotlib.pyplot as plt

car_prices = df['Price_in_thousands']

# 1. Descriptive statistics for car prices
print("Descriptive statistics for car prices:")
car_prices.describe()

# 2. Histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(car_prices, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# 3. Box plot to identify the spread and any potential outliers
plt.figure(figsize=(8, 5))
plt.boxplot(car_prices, vert=False, patch_artist=True)
plt.title('Box Plot of Car Prices')
plt.xlabel('Price (in thousands)')
plt.show()
```

###INSIGHT

Car Prices:The average car price is around 27,000 (in thousands), with some very high-priced models skewing the distribution.

The box plot and histogram showed a right-skewed distribution, indicating that most cars are priced in the lower range with a few very expensive models driving up the mean.

####QUSETION 15

Plot a correlation matrix to see if different numerical variables in the dataset correlate with each other

```python

import seaborn as sns

# create the correlation matrix
numerical_cols = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_cols.corr()

# plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

###INSIGHT

Correlation and Relationships:
Correlation Matrix:
Positive correlations were observed between engine size and horsepower, as expected (larger engines tend to have more power).

Price correlated positively with horsepower and curb weight, indicating that more powerful and heavier cars tend to be more expensive.
Fuel efficiency showed a negative correlation with engine size and curb weight, implying that larger and heavier cars tend to be less fuel-efficient.
